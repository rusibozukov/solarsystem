```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure and an axes
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_facecolor('black') # Space background

# Add random stars to the background
num_stars = 100
star_x = np.random.uniform(-4, 4, num_stars)
star_y = np.random.uniform(-4, 4, num_stars)
star_sizes = np.random.uniform(1, 8, num_stars)
ax.scatter(star_x, star_y, s=star_sizes, color='white', alpha=0.8, zorder=1)
# Attempt to remove the previously plotted white stars.
# This assumes the stars were the last collection added to the axes.
if ax.collections:
    ax.collections[-1].remove()

# Update star sizes to be smaller and random
# Original sizes were np.random.uniform(1, 8, num_stars)
updated_star_sizes = np.random.uniform(0.5, 3.5, num_stars)

# Update star colors to vary from pure white to pale yellow
# Colors will be (1, 1, B) where B is between 0.7 (pale yellow) and 1.0 (white)
r_channel = np.ones(num_stars)
g_channel = np.ones(num_stars)
b_channel = np.random.uniform(0.7, 1.0, num_stars)
updated_star_colors = np.stack([r_channel, g_channel, b_channel], axis=-1)

# Plot the stars with updated properties
# star_x, star_y, num_stars are used from the scope above
ax.scatter(star_x, star_y, s=updated_star_sizes, c=updated_star_colors, alpha=0.8, zorder=1)

# Central Star
star_radius = 0.6
star = plt.Circle((0, 0), star_radius, color='yellow', zorder=10)
ax.add_artist(star)

# Planets and their orbits
radius_planet_anna = 0.2

# Daniel ist jnger als Anna, daher ist sein Radius kleiner.
radius_planet_daniel = radius_planet_anna * 0.5

# Gabriel ist jnger als Daniel, daher ist sein Radius noch kleiner.
radius_planet_gabriel = radius_planet_daniel * 0.8

# Define the planet_data list with these updated radii.
# Other properties (orbit_radius, color, name) are kept as per the original structure
# implied by the problem description.
# This definition will be used by the subsequent loop if the original hardcoded
# definition of planet_data that follows this placeholder is removed or commented out by the user.
planet_data = [
    {'orbit_radius': 1.5, 'planet_radius': radius_planet_anna, 'color': 'blue', 'name': 'Anna'},
    {'orbit_radius': 2.5, 'planet_radius': radius_planet_daniel, 'color': 'red', 'name': 'Daniel'},
    {'orbit_radius': 3.5, 'planet_radius': radius_planet_gabriel, 'color': 'green', 'name': 'Gabriel'}
]

for planet_info in planet_data:
    # Orbit
    orbit = plt.Circle((0, 0), planet_info['orbit_radius'], color='gray', fill=False, linestyle='--', linewidth=0.5)
    ax.add_artist(orbit)

    # Planet position (random angle on the orbit)
    angle = np.random.rand() * 2 * np.pi
    planet_x = planet_info['orbit_radius'] * np.cos(angle)
    planet_y = planet_info['orbit_radius'] * np.sin(angle)

    # Planet
    planet = plt.Circle((planet_x, planet_y), planet_info['planet_radius'], color=planet_info['color'], zorder=5)
    ax.add_artist(planet)

# Set plot limits and aspect ratio
max_orbit = max(p['orbit_radius'] for p in planet_data) + 0.5
ax.set_xlim(-max_orbit, max_orbit)
ax.set_ylim(-max_orbit, max_orbit)
ax.set_aspect('equal', adjustable='box')

# Remove axes
# ax.axis('off')

# plt.title("Imaginary Solar System", color='black', fontsize=16)
plt.show()
```
